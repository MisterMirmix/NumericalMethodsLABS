```python
# Входной файл
# а, в – пределы интегрирования;
# ε точность:
# N - начальное число частичных отрезков:
# Hmin - наименьшее допустимое значение шага интегрирования.
# Выходной файл
# IER код завершения:
# IER=0 - решение получено:
# IER=1 решение не получено, погрешность перестала уменьшаться.
# IER=2 решение не получено, шаг интегрирования стал недопустимо малым.
# IER=3 ошибка входных данных ba
# Ih вычисленное значение интеграла.
# Ерѕ точность полученного решения.
# Н-шаг. при котором было получено решение
```

```python
import numpy as np
# a,b = (0.,1.)
# E = 0.000001
# E = 2**(-52)
# N = 4
# Hmin = 0.0000000000001
# print(f'границы отрезков = {a,b}')
# print(f'погрешность = {E}')
# print(f'минимальный шаг = {Hmin}')
# print(f'количество отрезков разбиения = {N}')

#подсчет интеграла методом трапеций с шагом с точность E и минимальным шагом Hmin
def I(a, b, E, N, Hmin, f):
    HminError = False
    EPSError = False
    
    h = (b-a)/N
    IhArray = np.array([Ih(a, b, N, f), Ih(a, b, 2*N, f)])
    EPS = epsilonRunge(IhArray[0],IhArray[1])
    print("------------------------------------------------------------")
    print(f'шаг h = {h}')
    print(f'интеграл Ih с шагом h = {IhArray[0]}')
    print(f'погрешность по правилу рунге = {EPS}')
    print("------------------------------------------------------------")
    if(EPS <= E):
        return IhArray[0],EPS,h,HminError,EPSError
    step = 0
    # E - EPS < 0 and Hmin - h < 0
    while not (HminError or EPSError or EPS <= E):
        prevEPS = EPS
        step += 1
        h = h/2
        IhArray = np.append(IhArray, Ih(a, b, (2**(step+1))*N, f))
        EPS = epsilonRunge(IhArray[step],IhArray[step+1])
        if (EPS >= prevEPS):
            EPSError = True
        if (Hmin > h):
            HminError = True
        print("------------------------------------------------------------")
        print(f'номер итерации = {step+1}')
        print(f'шаг h = {h}')
        print(f'интеграл Ih с шагом h = {IhArray[step]}')
        print(f'погрешность по правилу рунге = {EPS}')
        print("------------------------------------------------------------")
    return IhArray[step],EPS,h,HminError,EPSError

#подсчет интеграла методом трапеций с шагом h = (b-a)/N
def Ih(a,b,N,f):
    h = (b-a)/N

    Ih = 0.5*(f(a)+f(b))
    for i in range(1,N):
        Ih += f(a+i*h)
    Ih = Ih*h
    
    # print(f'шаг h = {h}')
    # print(f'интеграл Ih с шагом h = {Ih}')
    return Ih

#подсчет погрешности по правилу рунге
def epsilonRunge(Ih,Ih2):
    EPSr = abs(Ih-Ih2)*4/3
    # print(f'погрешность по правилу рунге = {EPSr}')
    return (abs(Ih-Ih2)*4/3)

def main(a, b, E, N, Hmin, f):
    IER = 0
    Ih = 0 
    EPS = 0
    H = 0
    if(a >= b):
        IER = 3
        return Ih,EPS,H,IER
    Ih,EPS,H,HminError,EPSError = I(a, b, E, N, Hmin, f)
    if(HminError):
        IER = 2
    if(EPSError):
        IER = 1
    return Ih,EPS,H,IER
    
def fconst(x):
    return 10.;
def fx(x):
    return float(x)
def fx2(x):
    return float(6*x*x)
def fsin(x):
    return float(np.sin(x))
def f7x23(x):
    return float(7*x+23)

# def fx2a(x):
#     return float((x-a)*(x-a))
    
fileinput = "input.txt"
fileoutput = 'output.txt'
with open(fileinput, 'r') as file:
    line = file.readline()   
    A,B = np.array(list(map(float, line.strip().split())))
    E = float(file.readline().strip())
    Hmin = float(file.readline().strip())
    N = int(file.readline().strip())
    print("---------------------------------------------")
    print(f'Входные данные:')
    print(f'границы отрезков = {A,B}')
    print(f'погрешность = {E}')
    print(f'минимальный шаг = {Hmin}')
    print(f'количество отрезков разбиения = {N}')
    print("---------------------------------------------")
    #ЗДЕСЬ МОЖНО МЕНЯТЬ ФУНКЦИЮ
    Ih,EPS,H,IER = main(A, B, E, N, Hmin, fx2)
    print("---------------------------------------------")
    print(f'Входные данные:')
    print(f'границы отрезков = {A,B}')
    print(f'погрешность = {E}')
    print(f'минимальный шаг = {Hmin}')
    print(f'количество отрезков разбиения = {N}')
    print("---------------------------------------------")
    print("IER код завершения:")
#     Оборотов Максим Владимирович
# Курс   3	Группа 3	 Преподаватель:
# Доцент	Шабунина З.А.
    if(IER == 0):
        
        print("IER=0 - решение получено:")
        print("------------------------------------------------------------")     
        print(f'вычисленное значение интеграла Ih = {Ih}')
        print(f'точность полученного решения Ерѕ = {EPS}')
        print(f'шаг, при котором было получено решение Н = {H}')
        print("------------------------------------------------------------")
    if(IER == 1):
        print("IER=1 решение не получено, погрешность перестала уменьшаться.")
    if(IER == 2):
        print("IER=2 решение не получено, шаг интегрирования стал недопустимо малым.")  
    if(IER == 3):
        print("IER=3 ошибка входных данных b <= a")
    with open(fileoutput, 'w') as fileOutput: 
            fileOutput.write("Input: \n")
            fileOutput.write(f"A, B: {A,B}\n")
            fileOutput.write(f"E: {E}\n")
            fileOutput.write(f"Hmin: {Hmin}\n")
            fileOutput.write(f"N: {N}\n")
            fileOutput.write("Output:\n")
            if(IER == 0):
                fileOutput.write("IER = 0;\n")
                fileOutput.write(f"Ih = {Ih}\n")
                fileOutput.write(f"ЕPS = {EPS}\n")
                fileOutput.write(f"H = {H}\n")
            if(IER == 1):
                fileOutput.write("IER = 1 Hmin > H;\n")
            if(IER == 2):
                fileOutput.write("IER = 2 new EPS equals previous EPS;\n")
            if(IER == 3):
                fileOutput.write("IER = 3 b >= a;\n")
```
