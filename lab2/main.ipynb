```python
%run Spline.ipynb

def main(X,Y,A,B,XX):
    IER = 0
    YY = 0
    if(X.shape[0]<2):
        IER = 1
    else:
        if not (isAscendingOrder(X)):
            IER = 2
        else:
            if not (pointInInterval(XX, X)):
                IER = 3
            else:
                print("---------------------------------------------")
                print("Трехточечная прогонка:")
                print("---------------------------------------------")
                # a,b,c,d = splineCoefficients(X,Y,A,B)
                a,b,c,d = splineCoefficients3(X,Y,A,B)
                print("---------------------------------------------")
                print(f'ai: {a}')
                print(f'bi: {b}')
                print(f'ci: {c}')
                print(f'di: {d}')
                print("---------------------------------------------")
                print("Сплайны на i-ом отрезке:")
                print("---------------------------------------------")
                for i in range(X.shape[0]-1):   
                    print(f'S{i+1}: {a[i]-b[i]*X[i+1]+X[i+1]*X[i+1]*c[i]/2-X[i+1]*X[i+1]*X[i+1]*d[i]/6} + {b[i]-c[i]*X[i+1]+d[i]*X[i+1]*X[i+1]/2}x + {c[i]/2-d[i]*X[i+1]/2}x\u00b2 + {d[i]/6}x\u00b3')       
                YY = functionDiff(XX, X, a,b,c,d)
                print(f"YY - значение функции в точке {XX} =", functionVal(XX, X, a,b,c,d))
                print("---------------------------------------------")
                
    return IER, YY
fileinput = "input.txt"
fileoutput = 'output.txt'
file_path = "Spline.txt"
with open(fileinput, 'r') as file:
    line = file.readline()
    while line:
        print("---------------------------------------------")
        
        X = np.array(list(map(float, line.strip().split())))
        line = file.readline()
        # Считываем Y-вектор (вторая строка)
        Y = np.array(list(map(float, line.strip().split())))
        line = file.readline()
        # Считываем количество узлов интерполирования (третья строка)
        A,B = np.array(list(map(float, line.strip().split())))
        line = file.readline()
        XX = list(map(float, line.strip().split()))
        line = file.readline()
        print("Входные параметры:")
        print("X-вектор:", X)
        print("Y-вектор:", Y)
        print("Граничные условия:", A, B)
        print("XX точка:", XX)
        
        IER,YY = main(X,Y,A,B,XX)
        print("Выходные параметры:")
        print("IER – индикатор ошибки:")
        if(IER == 0):
            print("IER = 0 – нет ошибки;")
            print(f"YY - значение первой прозводной в точке {XX} =", YY)
        if(IER == 1):
            print("IER = 1 – кубический сплайн не может быть построен ( N < 2 );")
        if(IER == 2):
            print("IER = 2 – нарушен порядок возрастания аргумента в входном векторе X;")
        if(IER == 3):
            print("IER = 3 – точка XX не принадлежит отрезку [x0 , xN];")
        with open(fileoutput, 'w') as fileOutput: 
            fileOutput.write("Входные параметры:\n")
            fileOutput.write(f"X-вектор: {X}\n")
            fileOutput.write(f"Y-вектор: {Y}\n")
            fileOutput.write(f"Граничные условия: {A, B}\n")
            fileOutput.write(f"XX точка: {XX}\n")
            fileOutput.write("Выходные параметры:\n")
            fileOutput.write("IER – индикатор ошибки:\n")
            if(IER == 0):
                fileOutput.write("IER = 0 – нет ошибки;\n")
                fileOutput.write(f"YY - значение первой прозводной в точке {XX} = {YY}\n")
            if(IER == 1):
                fileOutput.write("IER = 1 – кубический сплайн не может быть построен ( N < 2 );\n")
            if(IER == 2):
                fileOutput.write("IER = 2 – нарушен порядок возрастания аргумента в входном векторе X;\n")
            if(IER == 3):
                fileOutput.write("IER = 3 – точка XX не принадлежит отрезку [x0 , xN];\n")
        print("---------------------------------------------")
        
# file_path = "TMA.txt"

# file_path = "inputTMA.txt"
# with open(file_path, 'r') as file:
#     line = file.readline()
#     while line:
#         print("---------------------------------------------")
#         b = np.array(list(map(float, line.strip().split())))
#         line = file.readline()
#         c = np.array(list(map(float, line.strip().split())))
#         line = file.readline()
#         a = np.array(list(map(float, line.strip().split())))
#         line = file.readline()
#         f = np.array(list(map(float, line.strip().split())))
#         line = file.readline()
        
#         isTMA = isTMAPossible(a,b,c)
#         print("---------------------------------------------")       
#         print(f'Bi = {b}')
#         print(f'Ci = {c}')
#         print(f'Ai = {a}')
#         print(f'Fi = {f}')
#         print(f'Достаточное условие МТП: {isTMA}')
#         print("---------------------------------------------")
#         if(isTMA):
#             print(f'МТП:')
#             x = TMA(a,b,c,f)
            
#             print("---------------------------------------------")
#         with open(fileoutput, 'w') as fileOutput: 
#             fileOutput.write(f'Bi = {b}\n')
#             fileOutput.write(f'Ci = {c}\n')
#             fileOutput.write(f'Ai = {a}\n')
#             fileOutput.write(f'Fi = {f}\n')
#             fileOutput.write(f'Достаточное условие МТП: {isTMA}\n')
#             fileOutput.write(f'Решение: {x}\n')
        
```
