```python
import numpy as np

#ТРЕХТОЧЕЧНАЯ ПРОГОНКА
def TMA(aArray, bArray, cArray, fvector, isShow = True):

    N = cArray.shape[0] - 1 

    muArray = np.zeros(N)
    nuArray = np.zeros(N)
    #deltaMU_i = Ci+Ai*MUi
    deltaMU = np.zeros(N-1)
    
    muArray[0] = -bArray[0]/cArray[0]
    nuArray[0] = fvector[0]/cArray[0]
    for i in range(0,N-1):
        deltaMU[i] = cArray[i+1] + muArray[i]*aArray[i]
        muArray[i+1] = -bArray[i+1]/(deltaMU[i])
        nuArray[i+1] = (fvector[i+1]-aArray[i]*nuArray[i] )/(deltaMU[i])
           
    xVector = np.zeros(N+1)
    
    xVector[N] = (fvector[N]-aArray[N-1]*nuArray[N-1])/(cArray[N]+aArray[N-1]*muArray[N-1])
    for i in range(N-1,-1,-1):
        xVector[i] = xVector[i+1]*muArray[i]+nuArray[i]
    
    if(isShow):     
        print(f'Bi = {bArray}')
        print(f'Ci = {cArray}')
        print(f'Ai = {aArray}')
        print(f'MUi = {muArray}')
        print(f'NUi = {nuArray}')
        print(f'deltaMU = {deltaMU}')
        print(f'Xi = {xVector}')
        
    return xVector

def isTMAPossible(a,b,c):
    isCorrect = True   
    N = c.shape[0] - 1            
    isCorrect = isCorrect and (abs(c[0]) >= abs(b[0]))   
    for i in range(1,N):
        isCorrect = isCorrect and (abs(c[i]) >= abs(a[i-1]) + abs(b[i]))
    isCorrect = isCorrect and (abs(c[N]) > abs(a[N-1])) 
    return isCorrect

#СПЛАЙН
def splineCoefficients3(X,Y,A,B):
    N = X.shape[0]-1
    aCoef = np.zeros(N)
    bCoef = np.zeros(N)
   
    dCoef = np.zeros(N)
   
    fVector = np.diff(Y)
    hVector = np.diff(X)
    
    cCoef = np.zeros(N+1)
    FVector = np.zeros(N+1)
        
    FVector[0] = A
    FVector[N] = B
    for i in range(1,N):
        FVector[i] = 6*(fVector[i]/hVector[i] - fVector[i-1]/hVector[i-1])
        
    print(f'f_(i+1)-f_(i): {fVector}')
    print(f'x_(i+1)-x_(i): {hVector}')
    print(f'F: {FVector}')

    aDiag = np.zeros(N)
    bDiag = np.zeros(N)
    cDiag = np.zeros(N+1)
    
    cDiag[0] = 1
    bDiag[0] = 0
    cDiag[N] = 1
    aDiag[N-1] = 0

    for i in range(1, N):
        aDiag[i-1] = hVector[i-1]
        cDiag[i] = 2*(hVector[i-1]+hVector[i])
        bDiag[i] = hVector[i]
       
    cCoef = TMA(aDiag, bDiag, cDiag, FVector)
    
    for i in range (N):
        aCoef[i] = Y[i+1]
        
    for i in range(0,N):
        dCoef[i] = (cCoef[i+1]-cCoef[i])
        dCoef[i] = dCoef[i] / hVector[i-1]

    for i in range(0,N):
        bCoef[i] = hVector[i]*cCoef[i+1]/2- hVector[i]*hVector[i]*dCoef[i]/6+fVector[i]/hVector[i]

    return aCoef,bCoef,cCoef[1:],dCoef

#ЗНАЧЕНИЕ ФУНКЦИИ
def functionVal(XX, axis, a,b,c,d):
    index = findIntervalIndex(XX, axis)
    YY =a[index] + b[index]*(XX-axis[index+1])+ c[index]*(XX-axis[index+1])*(XX-axis[index+1])/2 + d[index]*(XX-axis[index+1])*(XX-axis[index+1])*(XX-axis[index+1])/6
    return YY
    
def findIntervalIndex(XX, axis):
    for i in range(1,axis.shape[0]):
        if (XX <= axis[i]):
            index = i
            break
    print(f'INDEX: {index}')
    return index - 1

def pointInInterval(XX, axis):
    if(axis[0] > XX or axis[-1] < XX):
        return False
    return True

def isAscendingOrder(vector): 
    if not np.all(np.diff(vector) > 0):
        return False
    return True
#ОСНОВНАЯ ПРОГРАММА
def main(X,Y,A,B,XX):
    IER = 0
    YY = 0
    if(X.shape[0]<2):
        IER = 1
    else:
        if not (isAscendingOrder(X)):
            IER = 2
        else:
            if not (pointInInterval(XX, X)):
                IER = 3
            else:
                print("---------------------------------------------")
                print("Трехточечная прогонка:")
                print("---------------------------------------------")
                a,b,c,d = splineCoefficients3(X,Y,A,B)
                print("---------------------------------------------")
                print(f'ai: {a}')
                print(f'bi: {b}')
                print(f'ci: {c}')
                print(f'di: {d}')
                print("---------------------------------------------")
                print("Сплайны на i-ом отрезке:")
                print("---------------------------------------------")
                for i in range(X.shape[0]-1):   
                    print(f'S{i+1}: {a[i]-b[i]*X[i+1]+X[i+1]*X[i+1]*c[i]/2-X[i+1]*X[i+1]*X[i+1]*d[i]/6} + {b[i]-c[i]*X[i+1]+d[i]*X[i+1]*X[i+1]/2}x + {c[i]/2-d[i]*X[i+1]/2}x\u00b2 + {d[i]/6}x\u00b3')       
                YY = functionVal(XX, X, a,b,c,d)
                print("---------------------------------------------")
                
    return IER, YY
fileinput = "input.txt"
fileoutput = 'output.txt'
file_path = "Spline.txt"
with open(fileinput, 'r') as file:
    line = file.readline()
    while line:
        print("---------------------------------------------")
        
        X = np.array(list(map(float, line.strip().split())))
        line = file.readline()
        # Считываем Y-вектор (вторая строка)
        Y = np.array(list(map(float, line.strip().split())))
        line = file.readline()
        # Считываем количество узлов интерполирования (третья строка)
        A,B = np.array(list(map(float, line.strip().split())))
        line = file.readline()
        XX = list(map(float, line.strip().split()))
        line = file.readline()
        print("Входные параметры:")
        print("X-вектор:", X)
        print("Y-вектор:", Y)
        print("Граничные условия:", A, B)
        print("XX точка:", XX)
        
        IER,YY = main(X,Y,A,B,XX)
        print("Выходные параметры:")
        print("IER – индикатор ошибки:")
        if(IER == 0):
            print("IER = 0 – нет ошибки;")
            print(f"YY - значение первой прозводной в точке {XX} =", YY)
        if(IER == 1):
            print("IER = 1 – кубический сплайн не может быть построен ( N < 2 );")
        if(IER == 2):
            print("IER = 2 – нарушен порядок возрастания аргумента в входном векторе X;")
        if(IER == 3):
            print("IER = 3 – точка XX не принадлежит отрезку [x0 , xN];")
        with open(fileoutput, 'w') as fileOutput: 
            fileOutput.write("Входные параметры:\n")
            fileOutput.write(f"X-вектор: {X}\n")
            fileOutput.write(f"Y-вектор: {Y}\n")
            fileOutput.write(f"Граничные условия: {A, B}\n")
            fileOutput.write(f"XX точка: {XX}\n")
            fileOutput.write("Выходные параметры:\n")
            fileOutput.write("IER – индикатор ошибки:\n")
            if(IER == 0):
                fileOutput.write("IER = 0 – нет ошибки;\n")
                fileOutput.write(f"YY - значение функции в точке {XX} = {YY}\n")
            if(IER == 1):
                fileOutput.write("IER = 1 – кубический сплайн не может быть построен ( N < 2 );\n")
            if(IER == 2):
                fileOutput.write("IER = 2 – нарушен порядок возрастания аргумента в входном векторе X;\n")
            if(IER == 3):
                fileOutput.write("IER = 3 – точка XX не принадлежит отрезку [x0 , xN];\n")
        print("---------------------------------------------")
        
# file_path = "TMA.txt"

# file_path = "inputTMA.txt"
# with open(file_path, 'r') as file:
#     line = file.readline()
#     while line:
#         print("---------------------------------------------")
#         b = np.array(list(map(float, line.strip().split())))
#         line = file.readline()
#         c = np.array(list(map(float, line.strip().split())))
#         line = file.readline()
#         a = np.array(list(map(float, line.strip().split())))
#         line = file.readline()
#         f = np.array(list(map(float, line.strip().split())))
#         line = file.readline()
        
#         isTMA = isTMAPossible(a,b,c)
#         print("---------------------------------------------")       
#         print(f'Bi = {b}')
#         print(f'Ci = {c}')
#         print(f'Ai = {a}')
#         print(f'Fi = {f}')
#         print(f'Достаточное условие МТП: {isTMA}')
#         print("---------------------------------------------")
#         if(isTMA):
#             print(f'МТП:')
#             x = TMA(a,b,c,f)
            
#             print("---------------------------------------------")
#         with open(fileoutput, 'w') as fileOutput: 
#             fileOutput.write(f'Bi = {b}\n')
#             fileOutput.write(f'Ci = {c}\n')
#             fileOutput.write(f'Ai = {a}\n')
#             fileOutput.write(f'Fi = {f}\n')
#             fileOutput.write(f'Достаточное условие МТП: {isTMA}\n')
#             fileOutput.write(f'Решение: {x}\n')
        
```
