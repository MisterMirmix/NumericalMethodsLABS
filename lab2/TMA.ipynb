```python
import numpy as np
import pandas as pd

def TMA(aArray, bArray, cArray, fvector, isShow = True):

    N = cArray.shape[0] - 1 

    muArray = np.zeros(N)
    nuArray = np.zeros(N)
    #deltaMU_i = Ci+Ai*MUi
    deltaMU = np.zeros(N-1)
    
    muArray[0] = -bArray[0]/cArray[0]
    nuArray[0] = fvector[0]/cArray[0]
    for i in range(0,N-1):
        deltaMU[i] = cArray[i+1] + muArray[i]*aArray[i]
        muArray[i+1] = -bArray[i+1]/(deltaMU[i])
        nuArray[i+1] = (fvector[i+1]-aArray[i]*nuArray[i] )/(deltaMU[i])
           
    xVector = np.zeros(N+1)
    
    xVector[N] = (fvector[N]-aArray[N-1]*nuArray[N-1])/(cArray[N]+aArray[N-1]*muArray[N-1])
    for i in range(N-1,-1,-1):
        xVector[i] = xVector[i+1]*muArray[i]+nuArray[i]
    
    if(isShow):
        print(f'Ai = {aArray}')
        print(f'Bi = {bArray}')
        print(f'Ci = {cArray}')
        print(f'MUi = {muArray}')
        print(f'NUi = {nuArray}')
        print(f'deltaMU = {deltaMU}')
        print(f'Xi = {xVector}')
        
    return xVector

def isTMAPossible(a,b,c):
    isCorrect = True   
    N = c.shape[0] - 1            
    isCorrect = isCorrect and (abs(c[0]) >= abs(b[0]))   
    for i in range(1,N):
        isCorrect = isCorrect and (abs(c[i]) >= abs(a[i-1]) + abs(b[i]))
    isCorrect = isCorrect and (abs(c[N]) > abs(a[N-1])) 
    return isCorrect
    

# matrix = np.arange(25).reshape(5,5)
# f1 = np.arange(matrix.shape[0])
# f2 = np.zeros(matrix.shape[0])
# matrix2 = makeThreeDiag(matrix)
# print("---------------------------------")
# print(matrix)
# print(f'МТП возомжен: {isTMAPossible(matrix)}')
# print(f'матрица трехдиагонального вида: {isThreeDiag(matrix)}')
# print("---------------------------------")
# print(matrix2)
# print(f'МТП возомжен: {isTMAPossible(matrix2)}')
# print(f'матрица трехдиагонального вида: {isThreeDiag(matrix2)}')
# for i in range(matrix2.shape[0]):
#     matrix2[i][i] = 50
# print("---------------------------------")
# print(matrix2)
# print(f'МТП возомжен: {isTMAPossible(matrix2)}')
# print(f'матрица трехдиагонального вида: {isThreeDiag(matrix2)}')
# print(f'вектор f: {f1}')
# TMA(matrix2,f1)
# print("---------------------------------")
# print(matrix2)
# print(f'вектор f: {f2}')
# TMA(matrix2,f2)
# print("---------------------------------")
# matrix2[0][4] = 8
# print(matrix2)
# print(f'МТП возомжен: {isTMAPossible(matrix2)}')
# print(f'матрица трехдиагонального вида: {isThreeDiag(matrix2)}')

# print("---------------------------------")
# matrix = np.array([[3, 1, 0], [-1, 4, 2], [0, 2, 3]])
# f = np.array([1,-3,5])
# print(f'вектор f: {f}')
# print(matrix)
# print(f'МТП возможен: {isTMAPossible(matrix)}')
# print(f'матрица трехдиагонального вида: {isThreeDiag(matrix)}')
# TMA(matrix,f)
# print("---------------------------------")
# matrix = np.zeros((10,10))
# f = np.array([1,14,14,14,14,14,14,14,14,1])
# nm = makeThreeDiag2(matrix)
# print(TMA(nm,f))
f = np.array([1,14,14,14,14,14,14,14,14,1])
b = np.array([0,1,1,1,1,1,1,1,1])
c = np.array([1,12,12,12,12,12,12,12,12,1])
a = np.array([1,1,1,1,1,1,1,1,0])
print(isTMAPossible(a,b,c))
print(TMA(a,b,c,f))
f = np.array([1,-3,5])
b = np.array([1,2])
c = np.array([3,4,3])
a = np.array([-1,2])
print(isTMAPossible(a,b,c))
print(TMA(a,b,c,f))
```
