```python
import numpy as np
import pandas as pd

def TMA(matrix, fvector, isShow = True):
    N = matrix.shape[0] - 1
    aArray = np.zeros(N)
    bArray = np.zeros(N)
    cArray = np.zeros(N+1)

    bArray[0] = matrix[0][1]
    cArray[0] = matrix[0][0]
    
    for i in range(1,N):
        aArray[i-1] = matrix[i][i-1]
        cArray[i] = matrix[i][i]
        bArray[i] = matrix[i][i+1]

    aArray[N-1] = matrix[N][N-1]
    cArray[N] = matrix[N][N]
    

    muArray = np.zeros(N)
    nuArray = np.zeros(N)
    #deltaMU_i = Ci+Ai*MUi
    deltaMU = np.zeros(N-1)
    
    muArray[0] = -bArray[0]/cArray[0]
    nuArray[0] = fvector[0]/cArray[0]
    for i in range(0,N-1):
        deltaMU[i] = cArray[i+1] + muArray[i]*aArray[i]
        muArray[i+1] = -bArray[i+1]/(deltaMU[i])
        nuArray[i+1] = (fvector[i+1]-aArray[i]*nuArray[i] )/(deltaMU[i])
           
    xVector = np.zeros(N+1)
    
    xVector[N] = (fvector[N]-aArray[N-1]*nuArray[N-1])/(cArray[N]+aArray[N-1]*muArray[N-1])
    for i in range(N-1,-1,-1):
        xVector[i] = xVector[i+1]*muArray[i]+nuArray[i]
    
    if(isShow):
        print(f'Ai = {aArray}')
        print(f'Bi = {bArray}')
        print(f'Ci = {cArray}')
        print(f'MUi = {muArray}')
        print(f'NUi = {nuArray}')
        print(f'deltaMU = {deltaMU}')
        print(f'Xi = {xVector}')
        
    return xVector
#проверка на трехдиагональность матрицы
def isThreeDiag(matrix):
    isCorrect = True
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            if (i != j and i + 1 != j and i - 1 != j):
                isCorrect = isCorrect and matrix[i][j] == 0
    return isCorrect
#создание новой трехдиагональность матрицы из данной матрицы    
def makeThreeDiag(matrix):
    newMatrix = np.zeros_like(matrix)
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            if not (i != j and i + 1 != j and i - 1 != j):
                newMatrix[i][j] = matrix[i][j]
    return newMatrix

def makeThreeDiag2(matrix):
    newMatrix = np.zeros_like(matrix)
    newMatrix[0][0] = 1
    newMatrix[matrix.shape[0]-1][matrix.shape[0]-1] = 1
    for i in range(1,matrix.shape[0]-1):
        newMatrix[i][i-1] = 1
        newMatrix[i][i] = 12
        newMatrix[i][i+1] = 1
    return newMatrix
#проверка достаточного условия применимости МТП
def isTMAPossible(matrix):
    isCorrect = True   
    N = matrix.shape[0] - 1            
    isCorrect = isCorrect and (abs(matrix[0][0]) >= abs(matrix[0][1]))   
    for i in range(1,N):
        isCorrect = isCorrect and (abs(matrix[i][i]) >= abs(matrix[i][i+1]) + abs(matrix[i][i-1]))
    isCorrect = isCorrect and (abs(matrix[N][N]) > abs(matrix[N][N-1])) 
    return isCorrect
    

# matrix = np.arange(25).reshape(5,5)
# f1 = np.arange(matrix.shape[0])
# f2 = np.zeros(matrix.shape[0])
# matrix2 = makeThreeDiag(matrix)
# print("---------------------------------")
# print(matrix)
# print(f'МТП возомжен: {isTMAPossible(matrix)}')
# print(f'матрица трехдиагонального вида: {isThreeDiag(matrix)}')
# print("---------------------------------")
# print(matrix2)
# print(f'МТП возомжен: {isTMAPossible(matrix2)}')
# print(f'матрица трехдиагонального вида: {isThreeDiag(matrix2)}')
# for i in range(matrix2.shape[0]):
#     matrix2[i][i] = 50
# print("---------------------------------")
# print(matrix2)
# print(f'МТП возомжен: {isTMAPossible(matrix2)}')
# print(f'матрица трехдиагонального вида: {isThreeDiag(matrix2)}')
# print(f'вектор f: {f1}')
# TMA(matrix2,f1)
# print("---------------------------------")
# print(matrix2)
# print(f'вектор f: {f2}')
# TMA(matrix2,f2)
# print("---------------------------------")
# matrix2[0][4] = 8
# print(matrix2)
# print(f'МТП возомжен: {isTMAPossible(matrix2)}')
# print(f'матрица трехдиагонального вида: {isThreeDiag(matrix2)}')

# print("---------------------------------")
# matrix = np.array([[3, 1, 0], [-1, 4, 2], [0, 2, 3]])
# f = np.array([1,-3,5])
# print(f'вектор f: {f}')
# print(matrix)
# print(f'МТП возможен: {isTMAPossible(matrix)}')
# print(f'матрица трехдиагонального вида: {isThreeDiag(matrix)}')
# TMA(matrix,f)
# print("---------------------------------")
# matrix = np.zeros((10,10))
# f = np.array([1,14,14,14,14,14,14,14,14,1])
# nm = makeThreeDiag2(matrix)
# print(TMA(nm,f))
```
