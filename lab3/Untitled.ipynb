```python

def solveLS(A,B,C,P,Q,F,X,k):
    # k = k - 1
    print(f'C: {C}')
    print(f'B: {B}')
    print(f'A: {A}')
    print(f'P: {P}')
    print(f'Q: {Q}')
    print(f'F: {F}')
        
    # print(mmm(A,B,C,P,Q))
    show_divided_differences(F, mmm(A,B,C,P,Q,k))
    
    C[0] = C[0]/B[0]
    P[0] = P[0]/B[0]
    Q[0] = Q[0]/B[0]
    F[0] = F[0]/B[0]
    B[0] = 1

    for i in range(1,k-1):   
        B[i] = B[i] - C[i-1]*A[i-1]
        Q[i] = Q[i] - Q[i-1]*A[i-1]
        P[i] = P[i] - P[i-1]*A[i-1]
        F[i] = F[i] - F[i-1]*A[i-1]
        A[i-1] = A[i-1] - B[i-1]*A[i-1]
        
        C[i] = C[i]/B[i]
        P[i] = P[i]/B[i]
        Q[i] = Q[i]/B[i]
        F[i] = F[i]/B[i]
        B[i] = 1

    i = k - 1
    B[i] = B[i] - C[i-1]*A[i-1]
    Q[i] = Q[i] - Q[i-1]*A[i-1]
    C[i] = C[i] - P[i-1]*A[i-1]
    P[i] = P[i] - P[i-1]*A[i-1]
    F[i] = F[i] - F[i-1]*A[i-1]
    A[i-1] = A[i-1] - B[i-1]*A[i-1]
        
    C[i] = C[i]/B[i]
    P[i] = P[i]/B[i]
    Q[i] = Q[i]/B[i]
    F[i] = F[i]/B[i]
    B[i] = 1

    i = k
    B[i] = B[i] - C[i-1]*A[i-1]
    Q[i] = Q[i] - Q[i-1]*A[i-1]
    C[i] = C[i] - Q[i-1]*A[i-1]
    P[i] = P[i] - P[i-1]*A[i-1]
    F[i] = F[i] - F[i-1]*A[i-1]
    A[i-1] = A[i-1] - B[i-1]*A[i-1]

    i = N-1
    A[i-1] = A[i-1]/B[i]
    P[i] = P[i]/B[i]
    Q[i] = Q[i]/B[i]
    F[i] = F[i]/B[i]
    B[i] = 1

    for i in range(N-2,k+2,-1):  
        B[i] = B[i] - A[i]*C[i]
        Q[i] = Q[i] - Q[i+1]*C[i]
        P[i] = P[i] - P[i+1]*C[i]
        F[i] = F[i] - F[i+1]*C[i]
        C[i] = C[i]-B[i+1]*C[i]
    
        A[i-1] = A[i-1]/B[i]
        P[i] = P[i]/B[i]
        Q[i] = Q[i]/B[i]
        F[i] = F[i]/B[i]
        B[i] = 1

    i = k+2
    B[i] = B[i] - A[i]*C[i]
    Q[i] = Q[i] - Q[i+1]*C[i]
    A[i-1] = A[i-1]- Q[i+1]*C[i]
    P[i] = P[i] - P[i+1]*C[i]
    F[i] = F[i] - F[i+1]*C[i]
    C[i] = C[i]-B[i+1]*C[i]
    
    A[i-1] = A[i-1]/B[i]
    P[i] = P[i]/B[i]
    Q[i] = Q[i]/B[i]
    F[i] = F[i]/B[i]
    B[i] = 1

    i = k+1
    B[i] = B[i] - A[i]*C[i]
    Q[i] = Q[i] - Q[i+1]*C[i]
    A[i-1] = A[i-1]- P[i+1]*C[i]
    P[i] = P[i] - P[i+1]*C[i]
    F[i] = F[i] - F[i+1]*C[i]
    C[i] = C[i]-B[i+1]*C[i]
    
    A[i-1] = A[i-1]/B[i]
    P[i] = P[i]/B[i]
    Q[i] = Q[i]/B[i]
    F[i] = F[i]/B[i]
    B[i] = 1

    i = k
    B[i] = B[i] - A[i]*C[i]
    Q[i] = Q[i] - Q[i+1]*C[i]
    P[i] = P[i] - P[i+1]*C[i]
    F[i] = F[i] - F[i+1]*C[i]
    C[i] = C[i]-B[i+1]*C[i]

    P[i] = P[i]/B[i]
    F[i] = F[i]/B[i]
    B[i] = 1



    Xres = np.zeros(N)

    Xres[k] = F[k]
    Xres[k+1] = F[k+1]-Xres[k]*P[k+1]
    Xres[k+2] = F[k+2]-Xres[k+1]*Q[k+2]-Xres[k]*P[k+2]
    for i in range(k+3,N):
        Xres[i] = F[i]-Xres[i-1]*A[i-1]-Xres[k+1]*Q[i]-Xres[k]*P[i]
    
    Xres[k-1] = F[k-1]-Xres[k+1]*Q[k-1]-Xres[k]*P[k-1]
    for i in range(k-2,-1,-1):
        Xres[i] = F[i]-Xres[i+1]*C[i]-Xres[k+1]*Q[i]-Xres[k]*P[i]
    print(f'C: {C}')
    print(f'B: {B}')
    print(f'A: {A}')
    print(f'P: {P}')
    print(f'Q: {Q}')
    print(f'F: {F}')
        
    # print(mmm(A,B,C,P,Q))
    
    show_divided_differences(F, mmm(A,B,C,P,Q,k))
    print(f'expected X: {X}')
    print(f'result X: {Xres}')
    
```
